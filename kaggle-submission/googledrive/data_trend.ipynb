```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
def getStates():
    ds = pd.read_csv('data/test.csv')
    states = ds['Province_State'][:50].values   
    return states
```

```python
states = getStates()
start = 0
```

```python
feature0 = 'Confirmed'

print(len(states))
for i in range(len(states)):
    state = states[i]
    print(state)
    ds = pd.read_csv('data/train.csv')
    ds = ds[ds['Province_State'] == state]
    value0 = ds[feature0].values
    print("confirmed:", value0[start:]) 
    val0Trend=value0[start:]
    plt.plot(val0Trend, label= state + ' Confirmed')
    plt.legend(loc='best')
    plt.xlabel('Day')
    plt.ylabel('Cases')
    plt.show()

```

```python
feature1 = 'Deaths'

print(len(states))
for i in range(len(states)):
    state = states[i]
    print(state)
    ds = pd.read_csv('data/train.csv')
    ds = ds[ds['Province_State'] == state]
    value1 = ds[feature1].values 
    print("deaths", value1[start:]) 
    val1Trend=value1[start:]
    plt.plot(val1Trend, label= state + ' Deaths')
    plt.legend(loc='best')
    plt.xlabel('Day')
    plt.ylabel('Cases')
    plt.show()
```
