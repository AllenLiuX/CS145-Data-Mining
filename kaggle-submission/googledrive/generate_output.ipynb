```python
import torch.nn as nn
import torch
import numpy as np
import torch.optim as optim
import pandas as pd
```

```python
def getStates():
    ds = pd.read_csv('data/test.csv')
    states = ds['Province_State'][:50].values
    return states
```

```python
input_size=1
hidden_size=16
num_layers=2
seq = 5
dropout = .1
seed = 4
states = getStates()
```

```python
class LSTM(nn.Module):
    def __init__(self):
        super(LSTM, self).__init__()
        self.lstm = nn.LSTM(input_size=input_size, 
                            hidden_size=hidden_size, 
                            num_layers=num_layers, 
                            batch_first=True, 
                            dropout=dropout)
        self.linear = nn.Linear(hidden_size * seq, 1)
    def forward(self, x):
        x, (h, c) = self.lstm(x)
        x = x.reshape(-1, hidden_size * seq)
        x = self.linear(x)
        return x
```

```python
for feature in ['Confirmed', 'Deaths']
    res = []
    for i in range(len(states)):
        state = states[i]
        print(i, states[i])

        ds = pd.read_csv('data/train.csv')
        ds = ds[ds['Province_State'] == state]
        value = ds[feature].values
        combine = [[value[i]] for i in range(len(value))]
        x = []
        y = []
        for i in range(len(combine)-seq-1):
            x.append(combine[i:i+seq])
            y.append(combine[i+seq][0])
        train_x = (torch.tensor(x).float()/100000.).reshape(-1, seq, input_size)
        train_y = (torch.tensor(y).float()/100000.).reshape(-1, 1)
        torch.manual_seed(seed)
        model = LSTM()
        optimzer = torch.optim.Adam(model.parameters(), lr=0.005)
        loss_func = nn.MSELoss()
        model.train()

        for epoch in range(2000):
            output = model(train_x)
            loss = loss_func(output, train_y)
            optimzer.zero_grad()
            loss.backward()
            optimzer.step()

        model.eval()

        x_test = []
        x_test.append(value[-seq:])

        predictions = []
        for i in range(26):
            x_pred = (torch.tensor(x_test[i]).float()/100000.).reshape(-1, seq, input_size)
            pred = int(model(x_pred) * 100000)
            x_append = x_test[i][1:].tolist()
            x_append.append(pred)
            x_test.append(np.array(x_append, dtype='int64'))
            predictions.append(pred)
        res.append(predictions)
    
    rerange = []
    for i in range(len(res[0])):
        for j in range(len(res)):
            rerange.append(res[j][i])
    df = pd.DataFrame(rerange)
    df.to_csv("{}.csv".format(feature),index=False,sep=',')
```
